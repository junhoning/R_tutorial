Q.1 Species 를 factor로 지정 해주세요

```r
data$Species <- as.factor(data$Species)
```

Q.2 species 중 setosa 만 있는 row 들을 뽑아주시고,  
그 중 Sepal.Width와 Petal.Length 만 보여주세요.  
(%in% 은 최소 한번은 사용 해주세요)

```r
select(filter(data, Species %in% 'setosa'), Sepal.Width, Petal.Length)
```

Q.3 species 중 setosa 만 있는 row 들을 뽑아주시고,  
그 중 Sepal.Width와 Petal.Length 만 보여주세요.  
(%>% 은 최소 한번은 사용 해주세요)

```r
data %>% select(Sepal.Width, Petal.Length) %>% filter(data$Species == 'setosa')
```

Q.4 Species 안에 있는 Sepal.Width의 평균 값을 구해주세요. 

```r
data %>% 
group_by(Species) %>% 
summarise(Average = mean(Sepal.Width, na.rm = TRUE))
```
